```python
from Bio import SeqIO

input_file = "NoGapsMSA_SIS1-on-top.fasta"
fasta_sequences = SeqIO.parse(open(input_file),'fasta')
```

```python
names, sequences = list(), list()
fastas = list()
for fasta in fasta_sequences:
    name, sequence = fasta.id, str(fasta.seq)
    print(fasta)
    names.append(name)
    sequences.append(sequence)
    fastas.append(fasta)
```

```python
unique_letters = ['V', 'W', 'H', 'G', 'T', 'Q', 'A', 'C', 'S', 'K', 'Y', 'L', 'R', 'D', 'M', 'E', 'I', 'F', 'N', 'P', '-']
PROTEIN_ENCODING = dict()
BINARY_ENCODING = dict()
for index, letter in enumerate(unique_letters):
    PROTEIN_ENCODING[letter] = [0 for _ in range(len(unique_letters))]
    PROTEIN_ENCODING[letter][index] = 1

    #inverse dictionary
    BINARY_ENCODING[str(PROTEIN_ENCODING[letter])] = letter
    
PROTEIN_ENCODING
```

```python
bin_seq = list()
for protein in fasta.seq:
    bin_seq.append(PROTEIN_ENCODING[protein])

print(bin_seq)
```

```python
seq = list()
for code in bin_seq:
    seq.append(BINARY_ENCODING[str(code)])
seq = "".join(seq)
print(seq)

if seq == fasta.seq:
    print("correct!")
```
